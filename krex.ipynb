```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('database.sqlite')

cur = conn.cursor()

query = '''
SELECT * 
FROM artists;
'''

cur.execute(query)

cur.description
```

```python
artists_data = cur.fetchall()

artists_df = pd.DataFrame(artists_data)

artists_df.columns = [i[0] for i in cur.description]
```

```python
artists_df.head()
```
